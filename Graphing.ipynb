```python
import matplotlib.pyplot as plt
```

```python
plt.plot(1,1,'bo',label='point A')
plt.plot(7,1,'ro',label='point B')
plt.plot(5,4,'go',label='point C')
plt.plot([1,7],[1,1],color=[.4,.2,.7],linewidth=2) # point to point lines
plt.plot([1,5],[1,4],color=[.4,.2,.7],linewidth=2)
plt.plot([7,5],[1,4],color=[.4,.2,.7],linewidth=2)
plt.plot([5,5],[1,4],color='k',linewidth=2)
plt.axis('square')
plt.xlabel('x')
plt.ylabel('y')
plt.axis([-.2,8,-.2,6])
plt.grid()
plt.legend()
plt.show()

```
